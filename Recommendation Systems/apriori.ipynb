```python
# Importing the libraries
import numpy as np
import pandas as pd
```

```python
# Data Preprocessing
data = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
```

```python
data
```

```python
tran=[]
for i in range(0,7500):
    tran.append([str(data.values[i,j]) for j in range(0,20)] )
```

```python
tran
```

```python
from apyori import apriori
```

```python
rules=apriori(tran,min_support=0.003,min_confidence=0.2,min_lift=2,max_length=4)
```

```python
results=list(rules)
```

```python
lift = []
association = []
for i in range (0, len(results)):
    lift.append(results[:len(results)][i][2][0][3])
    association.append(list(results[:len(results)][i][0]))
```

```python
rank = pd.DataFrame([association, lift]).T
rank.columns = ['Association', 'Lift']
```

```python
rank.sort_values('Lift', ascending=False).head(10)
```
